# Ejercicio

Crear un *programa* en **Python** en el que se grafique la funci√≥n $f\left(x\right)=x^{2n}\sin\left(\pi x\right)$ para $n$ valores en el intervalo para $x\in\left(0,1\right)$

```python
# Importar las librerias necesarias
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Definir la funci√≥n a graficar
def funcion(x,n):
    return x**(2*n)*np.sin(np.pi*x)

# Crear el dominio de graficaci√≥n
x = np.linspace(0,1,500)

# Crear un ciclo para graficar las primeras 10 funciones
for n in range(0, 30):
    
    # Evaluar la funci√≥n
    y = funcion(n, x)

    # Graficar la funci√≥n
    plt.plot(x,y)

# Configurar y Mostrar la gr√°fica
plt.grid(linestyle='--')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Gr√°fica de la funci√≥n $f\\left(x\\right)=x^{2n}\\sin\\left(\\pi x\\right)$ para los primeros valores de $n$')
plt.show()
```
