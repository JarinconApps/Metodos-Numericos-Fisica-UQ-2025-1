# Ejercicio

Crea un programa en Python que grafique la sucesión de términos para $$s\left(n\right)=\frac{2n\left(2n-1\right)}{\pi^{2}}$$ para los primeros 100 valores de $n$

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definir la sucesión
def sucesion(n):
    pi2 = np.pi**2
    return (2*n*(2*n-1)) / (pi2)

# Definir el dominio
n = np.linspace(1, 100, 100)

# Evaluar la sucesión
xn = sucesion(n)

# Graficar la sucesión
plt.scatter(n,xn)

# Configurar y mostrar la gráfica
plt.grid(linestyle='--')
plt.xlabel('$n$')
plt.ylabel('$\\left\\{ x_{n}\\right\\} _{n}$')
plt.title('Gráfica de la sucesión $\\frac{2n\\left(2n-1\\right)}{\\pi^{2}}$ para los primeros 100 valores de $n$')
plt.show()
```
