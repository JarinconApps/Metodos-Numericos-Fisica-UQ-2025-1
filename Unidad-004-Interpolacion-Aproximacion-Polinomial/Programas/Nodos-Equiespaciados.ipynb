### ***Métodos:*** Nodos Equiespaciados
En este módulo se define el método de los nodos equiespaciados usando el método de las diferencias divididas de Newton.

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt

from moduloInterpolacion import evalPoly
from moduloInterpolacion import NodosEquiespaciados
```

#### ***Probar el Método:*** NodosEquiespaciados
Probar el método de **NodosEquiespaciados** con la función seno con 24 nodos desde 0 hasta $2\pi$

```python
# Define una función que genera valores equiespaciados
def funcion(x):
    return np.sin(x)

# Crear los nodos equiespaciados
x = np.linspace(0, 2*np.pi, 24)

# Calcular el valor del h
h = x[1]-x[0]

# Calcular los valores de y
y_nodos = funcion(x)

# Usar el método NodosEquiespaciados para calcular el polinomio interpolador
poly = NodosEquiespaciados(y_nodos, x[0], h)
sPoly, pPoly = poly.popitem()

# Mostrar los coeficientes del polinomio interpolador
print(pPoly)

# Graficar el polinomio interpolador y los nodos
y_Poly = evalPoly(pPoly, x)
plt.plot(x, y_Poly)
plt.scatter(x, y_nodos, c='red', zorder=3)
plt.grid(linestyle='--')
plt.title('Gráfica del Polinomio Interpolador y los Nodos')
plt.xlabel('Valores de $x$')
plt.ylabel('Valores de $y$')

# Guardar y mostrar la gráfica
plt.savefig('../Interpolacion-Aproximacion-Polinomial/Imagenes/Ejemplo-Nodos-Equiespaciados.pdf')
plt.show()
```
