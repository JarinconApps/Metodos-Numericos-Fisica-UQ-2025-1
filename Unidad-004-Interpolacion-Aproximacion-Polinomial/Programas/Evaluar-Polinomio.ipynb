### ***Función:*** evalPoly
- **Descripcion:** Evalua un polinomio dado una lista de puntos y sus coeficientes
- **Parámetros:**
    - *coeficientes:* Lista de coeficientes del polinomio de la forma $[a_n, a_{n-1}, \ldots, a_1, a_0]$
    - *dominio:* Lista de valores a evaluar en el polinomio
    - *console:* Valor booleando. Permite mostrar los mensajes de procesos
- **Valor de Retorno:** Una lista de imagenes del dominio evaluado en el polinomio

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definir la función evalPoly
def evalPoly(coeficientes, dominio, console=False):

    """
    # ***Función:*** evalPoly
    - **Descripcion:** Evalua un polinomio dado una lista de puntos y sus coeficientes
    - **Parámetros:**
        - *coeficientes:* Lista de coeficientes del polinomio de la forma $[a_n, a_{n-1}, \\ldots, a_1, a_0]$
        - *dominio:* Lista de valores a evaluar en el polinomio
        - *console:* Valor booleando. Permite mostrar los mensajes de procesos
    - **Valor de Retorno:** Una lista de imagenes del dominio evaluado en el polinomio
    """

    # Obtener el grado del polinomio
    grado = np.shape(coeficientes)[0] - 1

    # Crear la imagen    
    imagen = np.empty((0, grado))
    dominio_ = np.array(dominio)
    
    # Crear un ciclo para evaluar el polinomio
    imagen = np.empty((0, grado))

    for x in dominio_:
        suma = 0
        for n,c in enumerate(coeficientes):
            suma = suma + c*x**(grado-n)

        imagen = np.append(imagen, suma)

    # Devolver la imagen
    return imagen
```

#### ***Pruebas***
Probar la función evalPoly con los coeficientes [   1,  -15,   85, -225,  274, -120] en el dominio [1, 2, 3, 4, 5]

```python
# Probar la función
coeficientes = [   1,  -15,   85, -225,  274, -120]
dominio = [1, 2, 3, 4, 5]
imagen = evalPoly(coeficientes, dominio)
print(imagen)
```

#### ***Gráfica del Polinomio***

Como ya tenemos una función para evaluar el polinomio dados sus coeficientes y un dominio, vamos a realizar su gráfica

```python
# Define los coeficientes del polinomio
coeficientes = [   1,  -15,   85, -225,  274, -120]

# Genera el dominio para graficar el polinomio
dominio = np.linspace(0.8, 5.2, 500)
imagen = evalPoly(coeficientes, dominio)
plt.plot(dominio, imagen)

# Genera un dominio discreto para evaluar los cortes/raíces
dominio_ = np.linspace(1, 5, 5)
imagen_ = evalPoly(coeficientes, dominio_)
plt.scatter(dominio_, imagen_, c='red')

# Configura la gráfica
plt.grid(linestyle='--')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.title('Gráfica del Polinomio Evaluado')

# Mostrar la gráfica
plt.show()
```
