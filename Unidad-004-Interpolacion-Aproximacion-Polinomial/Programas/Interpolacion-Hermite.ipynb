Calcular el polinomio intepolador de Hermite de los nodos $x=[0, 1, 3, 5]$ para la función $f(x)=x\cdot\sin(\frac{\pi}{2}x)$

```python
# Definir los nodos de x, y y la derivada
nodos_x = np.array([0, 1, 3, 5])
nodos_y = np.array([0, 1, -3, 5])
nodos_dy = np.array([0, 1, -1, 1])

# Crear la Tupla para usar los polinomios de Hermite
nodos = (nodos_x, nodos_y, nodos_dy)
poly = Hermite(nodos)
print("Polinomio Interpolador: ", poly)

# Crear un dominio de graficación, evaluar la función y el polinomio calculado
x = np.linspace(0, 5, 100)
P = evalPoly(poly, x)
y = funcion(x)

# Realizar la gráfica de la función
plt.plot(x, y)
plt.plot(x, P)
plt.scatter(nodos_x, nodos_y, c='red', zorder=4)

# Configurar la gráfica
plt.title('Gráfica del Polinomio Interpolador')
plt.xlabel('Valores de $x$')
plt.ylabel('Valores de $y$')
plt.grid(linestyle='--')
plt.legend(["$y=f(x)$", "$y=P(x)$"])

# Guardar el gráfico y mostrar
plt.savefig('../Interpolacion-Aproximacion-Polinomial/Imagenes/Polinomio-Interpolador-Hermite.pdf')
plt.show()
```
