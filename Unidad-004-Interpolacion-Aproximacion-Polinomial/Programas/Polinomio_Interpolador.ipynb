### ***Polinomio Interpolador***

En este código se calcula un polinomio interpolador usando la forma más primitiva y es resolviendo el siguiente sistema de ecuaciones $$S=\begin{cases}
a_{3}+a_{2}+a_{1}+a_{0} & =1.32\\
8a_{3}+4a_{2}+2a_{1}+a_{0} & =1.75\\
27a_{3}+9a_{2}+3a_{1}+a_{0} & =1.94\\
64a_{3}+16a_{2}+4a_{1}+a_{0} & =1.84
\end{cases}$$ para ello se usa el módulo de **Numpy**: *np.linalg.solve(a,b.T)*

```python
# Importar Numpy
import numpy as np

# Crear la matriz de coeficientes
a = np.array([
    [1,1,1,1],
    [8, 4, 2, 1],
    [27, 9, 3, 1],
    [64, 16, 4, 1]
])

# Crear el vector de términos indepdientes
b = np.array([1.32, 1.75, 1.94, 1.84])

# Resolver el sistema
x = x = np.linalg.solve(a,b.T)

# Mostrar el resultado
print("Coeficientes del Polinomio: ", x)
```

#### ***Gráfica de la Función Polinómica***

Se define una función $Pn(x)$ dados los coeficientes en el código anterior, con estos se realiza un gráfico del polinomio interpolador (usando plt.plot) y de los puntos de interpolación (usando plt.scatter)

```python
# Importar MatPlotLib
import matplotlib.pyplot as plt

def pn(x):
    return -0.00833333*x**3-0.07*x**2+0.69833333*x+0.7

# Crear un vector de términos de x
x = np.linspace(0.9, 4.1, 50)

# Evaluar la función en los valores de x
y = pn(x)

# Graficar la función
plt.plot(x,y)

# Configurar el gráfico
plt.xlabel('Valores de $x$')
plt.ylabel('$y = P_n(x)$')
plt.title('Gráfica del Polinomio Interpolador')
plt.grid(linestyle='--')

# Graficar puntos
x1 = np.array([1,2,3,4])
y1 = pn(x1)
plt.scatter(x1,y1, c='red')

# Guardar el gráfico en formato PDF
plt.savefig('../Interpolacion-Aproximacion-Polinomial/Imagenes/grafica-polinomio-interpolador.pdf')

# Mostrar la gráfica
plt.show()

# Mostrar la aproximación de los puntos
print("Valores de los puntos de interpolación mediante el polinomio interpolador: \n", y1)
```
