### ***Ejercicio:*** Polinomio Interpolador

El objetivo de este ejercicio es usar de forma organizada los métodos (funciones) que se han creado hasta el momento sobre interpolación. Se ha definicio un archivo llamado `moduloInterpolacion.py` que contiene las definiciones de `MultiplicacionSintetica`, `Primos`, `evalPoly`, `polyInterpoLagrange` que permiten de forma más eficiente resolver el problema que nos ocupa en éste archivo, encontrar el polinomio interpolador.

El ejercicio busca generar los números primos en el intervalo $[2,40]$ y asociarles luego un número natural $\mathbb{N}$ a cada primo. Con estos nodos generar el polinomio interpolador usando para ellos los **Polinomios de Lagrange**. Finalmente se evalua el polinimio encontrado, se grafica y se guarda la gráfica en formato PDF.

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt

# Importar mis funciones personalizadas
from moduloInterpolacion import Primos
from moduloInterpolacion import polyInterpoLagrange
from moduloInterpolacion import evalPoly

# Definir la ruta de las imágenes
rutaImagenes = '../Interpolacion-Aproximacion-Polinomial/Imagenes/'
```

```python
# Definir los nodos
nodos_y = Primos(2, 40)
print(nodos_y)
nMax = len(nodos_y)
nodos_x = np.array([n+1 for n in range(nMax)])
print(nodos_x)
nodos = (nodos_x, nodos_y)

# Usar la función de interpolación
poly = polyInterpoLagrange(nodos)
print(poly)

# Crear el dominio de graficación y evaluar
x = np.linspace(1, nMax, 200)
y = evalPoly(poly, x)

# Graficamos el polinomio y los puntos
plt.plot(x, y)
plt.scatter(nodos_x, nodos_y, c='red', zorder=3)

# Configurar la gráfica
plt.grid(linestyle='--')
plt.xlabel('Números Naturales')
plt.ylabel('Números Primos')
plt.title('Gráfica del Polinomio Intepolador de Primeros '+str(nMax)+' Números Primos')

# Guardar la gráfica en formato PDF
plt.savefig(rutaImagenes + 'polinomio-numeros-primos.pdf')

# Mostrar la gráfica
plt.show()
```
