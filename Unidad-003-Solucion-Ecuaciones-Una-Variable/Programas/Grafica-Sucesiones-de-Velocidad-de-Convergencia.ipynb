# Ejercicio 1
Considere la sucesión anterior $\left(x_{n}\right)_{n}=\left(\frac{1}{3\sqrt[3]{4}}\right)^{n}$. Desarrolle un programa en Python que grafique esta sucesión para los primeros $n$ valores

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Graficar la sucesión del ejemplo: (1/(3*4**(1/3)))**n

# Construir la lista de puntos para n
n = np.linspace(0, 20, 21);
x = np.linspace(0, 20, 500);

# Definirla sucesión
def xn(n):
  return (1/(3*4**(1/3)))**n

# Graficar la sucesión
plt.plot(x, xn(x))
plt.scatter(n, xn(n), c='#E97203')
plt.xlabel('n')
plt.ylabel('x(n)')
plt.xticks(np.arange(0, 21, 2))
plt.grid(ls='--')
plt.title('Análisis de la Velocidad de Convergencia')

# Definir la leyedan para la gráfica
plt.legend(['Sucesión', '0.21^n'])
plt.show()
```
