# Descripción Gráfica del Método de Newton Rapson

Este programa tiene como objetivo ilustrar gráficamente el funcionamiento del método de Newton-Raphson. Para ello, se define la función $f(x) = e^{-x}-\frac{1}{2}\ln{x}$, y se establece un valor inicial $x_0 = 1$.

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definir el dominio de graficación
x = np.linspace(0.75, 2, 500)

# Definir la función y su derivada
def f(x):
    return np.exp(-x) - 0.5*np.log(x)

def df_dx(x):
    return -np.exp(-x)-1/(2*x)

def rt(x, m, x_1, y_1):
    return m*(x-x_1) + y_1

# Definir el tamaño de la gráfica
wg = 40 / 2.54
hg = 24 / 2.54
plt.figure(figsize=(wg, hg))

# Graficar la función
plt.plot(x, f(x), linewidth=1.5)

# Definimos el valor x_0, calculamos la derivada y la recta tangente

x_0 = 1
for i in range(1,4):
    y_0 = f(x_0)
    m = df_dx(x_0)
    plt.plot(x, rt(x, m, x_0, y_0), linewidth=1.25)
    plt.axvline(x=x_0, color='red', linestyle='--', linewidth=1.0)
    plt.plot(x_0, y_0, 'bo')

    # Calcular el nuevo x_0
    x_0 = (m*x_0-y_0) / m

# Configurar y mostrar la gráfica
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Descripció Gráfica del Método de Newton-Rapson")
plt.grid(linestyle='--')
plt.show()

```
