# Ejercicio
Crear un programa en Python que grafique las funciones $g_{1}\left(x\right)=x^{3}+1$, $f\left(x\right)=x$ y $g_{2}\left(x\right)=\sqrt[3]{x-1}$

```python
# Importar las librerias
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# Definir las funciones
def g1(x):
    return x**3 + 1

def f(x):
    return x

def g2(x):
  return np.cbrt(x-1)
```

```python
# Graficar las funciones en el intervalo (-2,2)

# Calcular el dominio
x = np.linspace(-2,-1,500)

# Graficar cada función
plt.plot(x,g1(x),label=r'$g_{1}(x) = x^3+1$')
plt.plot(x,f(x),label=r'$f(x)$')
plt.plot(x,g2(x),label=r'$g_{2}(x) = \sqrt[3]{x-1}$')

# Etiquetar los ejes del gráfico
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Activar el grid de la gráfica
plt.grid()

# Establecer el título del gráfico
plt.title('Gráfica Funciones: Método del Punto Fijo')

# Cambiar el rango de la gráfica de -4 a 4
plt.ylim(-7,1)

# Mostrar el gráfico
plt.show()

```
