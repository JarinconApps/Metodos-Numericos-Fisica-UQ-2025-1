# Ejercicio

Crear un *programa* en **Python** que grafique el polinomio $P_{3}\left(x\right)=x^{3}-10x^{2}+6x-30$ para $x\in\left[1,6\right]$

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definir el polinomio
def poly(x):
    return x**3 - 10*x**2 + 31*x - 30

# Crear el dominio de graficación
x = np.linspace(1, 6, 100)

# Evaluar el polinomio en x
y = poly(x)

# Graficar y configurar la gráfica
plt.plot(x, y)
plt.xlabel('$x$')
plt.ylabel('$P_n(x)$')
plt.title('Gráfico del Polinomio $P_n(x)$')
plt.grid(linestyle='--')
plt.show()
```
